## 2.3- Cargamos el fichero de gastos

```python
def Carga_1_archivoGasto(Origen):
    df_gastos=pd.read_fwf(Origen,
               header=None,
               names = ['ANOENC','NUMERO','CODIGO','GASTO','PORCENDES',
                        'PORCENIMP','CANTIDAD','GASTMON','GASTNOM1','GASTNOM2',
                        'GASTNOM3','GASTNOM4','GASTNOM5','FACTOR'],
               widths=[4,5,5,15,5,
                      5,12,15,13,13,
                      13,13,13,11]
               #colspecs=[(0,5),(5,9)]
              )
    return df_gastos
```

```python
def CreaDFgastos(RutaOrigen,FicheroOrigen):
    Origen=RutaOrigen+'\\'+FicheroOrigen
    df_gastos=Carga_1_archivoGasto(Origen)
    return df_gastos
```

```python
def AnexaDFgastos(df_gastos_total,RutaOrigen,FicheroOrigen):
    Origen=RutaOrigen+'\\'+FicheroOrigen
    df_gastos=Carga_1_archivoGasto(Origen)
    df_gastos_total=pd.concat([df_gastos_total,df_gastos],ignore_index=True)
    return df_gastos_total
```
